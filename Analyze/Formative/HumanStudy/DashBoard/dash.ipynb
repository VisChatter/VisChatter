```python
from PIL import Image, ImageDraw, ImageFont
def resize_image(img_path, max_width, max_height):
	img = Image.open(img_path)
	width, height = img.size

	# Check if height exceeds max_height
	if height > max_height:
		# Calculate new width to maintain aspect ratio
		new_width = int((max_height / height) * width)
		img = img.resize((new_width, max_height))
		width, height = img.size  # Update width and height after resize

	# Check if width exceeds max_width (though your current logic already handles this)

	return img
def create_dashboard(row,index):

	img1 = resize_image(row['Image1'], max_width=1600, max_height=600)
	img2 = resize_image(row['Image2'], max_width=1600, max_height=600)
 	# Determine the width and height of both images
	width1, height1 = img1.size
	width2, height2 = img2.size

	# Compare the widths to identify the larger image
	#TODO:ADD HEIGHT
	if width1 > width2:
		larger_img = img1
		smaller_img = img2
		larger_width = width1
		smaller_width = width2
	else:
		larger_img = img2
		smaller_img = img1
		larger_width = width2
		smaller_width = width1

	if larger_width + smaller_width > 1600:
		# Resize the larger image to fit within the limit
		new_larger_width = 1600 - smaller_width
		new_larger_height = int((new_larger_width / larger_width) * larger_img.height)
		larger_img = larger_img.resize((new_larger_width, new_larger_height))

	# Create a new image for the dashboard
	dashboard = Image.new('RGB', (1600, 850), (255, 255, 255))
	
	# Draw metrics
	draw = ImageDraw.Draw(dashboard)
	font = ImageFont.load_default(size=30)
	font_large = ImageFont.load_default(size=64)
	# fontsize=3200

	draw.text((50, 30), f"{list(fact[index].keys())[0]}", font=font, fill=(0, 0, 0))
	draw.text((550, 30), f"{list(fact[index].keys())[1]}", font=font, fill=(0, 0, 0))
	draw.text((1100, 30), f"{list(fact[index].keys())[2]}", font=font, fill=(0, 0, 0))
	draw.text((50, 60), f"{list(fact[index].values())[0]}", font=font_large, fill=(0, 0, 0))
	draw.text((550, 60), f"{list(fact[index].values())[1]}", font=font_large, fill=(0, 0, 0))
	draw.text((1100, 60), f"{list(fact[index].values())[2]}", font=font_large, fill=(0, 0, 0))   
	# Paste images onto the dashboard
   
	dashboard.paste(smaller_img, (0, 250))
	if larger_width + smaller_width > 1500:
		dashboard.paste(larger_img, (smaller_width, 250))
	else:
		dashboard.paste(larger_img, (smaller_width+100, 250))
	
	output_path = f"{row['Image1'][3:-4]}.png"
	# Save the dashboard as a PNG file
	dashboard.save(output_path)

# Example usage
for index, row in df.iterrows():
	create_dashboard(row,index)
```
