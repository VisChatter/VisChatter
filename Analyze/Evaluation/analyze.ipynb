```python
import pandas as pd

# Read the CSV file into a DataFrame
df = pd.read_csv('AllData.csv')

# List of substrings to filter columns
substrings = ["_1_x", "_1_y", "_2_x", "_2_y"]

# Filter columns that contain any of the substrings
filtered_columns = [col for col in df.columns if any(sub in col for sub in substrings)]

# Create a new DataFrame with the filtered columns
new_df = df[filtered_columns]
# .fillna(0,inplace=True)
# Display the new DataFrame
new_df=new_df.fillna(0)

# Optionally, save the new DataFrame to a new CSV file
# new_df.to_csv('filtered_columns.csv', index=False)
```

```python
import pandas as pd
from PIL import Image, ImageDraw

# Load your CSV file
# df = pd.read_csv('Merge.csv')

# Load the image
image_path = 'VisChatter/ComplexLine2.png'
image = Image.open(image_path).convert("RGBA")
original_width, original_height = image.size
new_width = 700
new_height = int((new_width / original_width) * original_height)
resized_image = image.resize((new_width, new_height), Image.Resampling.LANCZOS)
overlay = Image.new("RGBA", resized_image.size, (255, 255, 255, 0))
draw = ImageDraw.Draw(overlay)

# Define the color and size of the dots
# dot_color1 = (255, 0, 0, 255)  # Stroke color for the first dot (red)
dot_color1 = (255, 0, 0, 255) # Stroke color for the second dot (green)
dot_radius = 10
stroke_width = 5  # Width of the stroke

# Iterate over each row in the DataFrame
for index, row in new_df.iterrows():
	if index > 2:
		# Get the positions of the first and second dots
		x1, y1 = row['Q26_1_x'], row['Q26_1_y']
		x2, y2 = row['Q26_2_x'], row['Q26_2_y']
		
		# Draw the first dot with transparent fill and colored stroke
		draw.ellipse(
			[(int(x1) - dot_radius, int(y1) - dot_radius), (int(x1) + dot_radius, int(y1) + dot_radius)],
			fill=(255, 255, 255, 0),  # Transparent fill
			outline=dot_color1,       # Stroke color
			width=stroke_width        # Stroke width
		)
		
		# Draw the second dot with transparent fill and colored stroke
		draw.ellipse(
			[(int(x2) - dot_radius, int(y2) - dot_radius), (int(x2) + dot_radius, int(y2) + dot_radius)],
			fill=(255, 255, 255, 0),  # Transparent fill
			outline=dot_color1,       # Stroke color
			width=stroke_width        # Stroke width
		)

# Combine the original image with the overlay
combined_image = Image.alpha_composite(resized_image, overlay)

# Save the modified image
combined_image.save('CompLine1.png')


image_path = 'VisChatter/ComplexLine1.png'
image = Image.open(image_path).convert("RGBA")
original_width, original_height = image.size

new_height = int((new_width / original_width) * original_height)
resized_image = image.resize((new_width, new_height), Image.Resampling.LANCZOS)
overlay = Image.new("RGBA", resized_image.size, (255, 255, 255, 0))
draw = ImageDraw.Draw(overlay)

for index, row in new_df.iterrows():
	if index>2:
	# Get the positions of the first and second dots
		x1, y1 = row['Q80_1_x'], row['Q80_1_y']
		x2, y2 = row['Q80_2_x'], row['Q80_2_y']
		
		draw.ellipse(
			[(int(x1) - dot_radius, int(y1) - dot_radius), (int(x1) + dot_radius, int(y1) + dot_radius)],
			fill=(255, 255, 255, 0),  # Transparent fill
			outline=dot_color1,       # Stroke color
			width=stroke_width        # Stroke width
		)
		
		# Draw the second dot with transparent fill and colored stroke
		draw.ellipse(
			[(int(x2) - dot_radius, int(y2) - dot_radius), (int(x2) + dot_radius, int(y2) + dot_radius)],
			fill=(255, 255, 255, 0),  # Transparent fill
			outline=dot_color1,       # Stroke color
			width=stroke_width        # Stroke width
		)

combined_image = Image.alpha_composite(resized_image, overlay)
# Save the modified image
combined_image.save('CompLine2.png')


image_path = 'VisChatter/ComplexLine2.png'
image = Image.open(image_path).convert("RGBA")
original_width, original_height = image.size

new_height = int((new_width / original_width) * original_height)
resized_image = image.resize((new_width, new_height), Image.Resampling.LANCZOS)
overlay = Image.new("RGBA", resized_image.size, (255, 255, 255, 0))
draw = ImageDraw.Draw(overlay)

for index, row in new_df.iterrows():
	if index>2:
	# Get the positions of the first and second dots
		x1, y1 = row['Q67_1_x'], row['Q67_1_y']
		x2, y2 = row['Q67_2_x'], row['Q67_2_y']
		
		draw.ellipse(
			[(int(x1) - dot_radius, int(y1) - dot_radius), (int(x1) + dot_radius, int(y1) + dot_radius)],
			fill=(255, 255, 255, 0),  # Transparent fill
			outline=dot_color1,       # Stroke color
			width=stroke_width        # Stroke width
		)
		
		# Draw the second dot with transparent fill and colored stroke
		draw.ellipse(
			[(int(x2) - dot_radius, int(y2) - dot_radius), (int(x2) + dot_radius, int(y2) + dot_radius)],
			fill=(255, 255, 255, 0),  # Transparent fill
			outline=dot_color1,       # Stroke color
			width=stroke_width        # Stroke width
		)

combined_image = Image.alpha_composite(resized_image, overlay)
# Save the modified image
combined_image.save('CompLine1-VC.png')


image_path = 'VisChatter/ComplexLine1.png'
image = Image.open(image_path).convert("RGBA")
original_width, original_height = image.size

new_height = int((new_width / original_width) * original_height)
resized_image = image.resize((new_width, new_height), Image.Resampling.LANCZOS)
overlay = Image.new("RGBA", resized_image.size, (255, 255, 255, 0))
draw = ImageDraw.Draw(overlay)

for index, row in new_df.iterrows():
	if index>2:
	# Get the positions of the first and second dots
		x1, y1 = row['Q100_1_x'], row['Q100_1_y']
		x2, y2 = row['Q100_2_x'], row['Q100_2_y']
		
		draw.ellipse(
			[(int(x1) - dot_radius, int(y1) - dot_radius), (int(x1) + dot_radius, int(y1) + dot_radius)],
			fill=(255, 255, 255, 0),  # Transparent fill
			outline=dot_color1,       # Stroke color
			width=stroke_width        # Stroke width
		)
		
		# Draw the second dot with transparent fill and colored stroke
		draw.ellipse(
			[(int(x2) - dot_radius, int(y2) - dot_radius), (int(x2) + dot_radius, int(y2) + dot_radius)],
			fill=(255, 255, 255, 0),  # Transparent fill
			outline=dot_color1,       # Stroke color
			width=stroke_width        # Stroke width
		)

combined_image = Image.alpha_composite(resized_image, overlay)
# Save the modified image
combined_image.save('CompLine2-VC.png')
```
